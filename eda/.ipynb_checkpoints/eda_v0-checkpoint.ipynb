```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
import seaborn as sns
```

```python
district_df = pd.read_csv("../data/district_case_comps.csv")
crime_df = pd.read_csv("../data/dstrIPC_1.csv")
```

```python
crime_df["DISTRICT"] = crime_df["DISTRICT"].str.lower()
```

```python
crime_df = crime_df.groupby(["STATE/UT", "DISTRICT"])["TOTAL IPC CRIMES"].mean().reset_index()
```

```python
merged_df = district_df.merge(crime_df, left_on="district_name", right_on="DISTRICT", how = "inner")
merged_df
```

```python
merged_df["crime_rate"] = merged_df["TOTAL IPC CRIMES"]/merged_df["tot_p"]
```

```python
analysis_df = merged_df[merged_df["crime_rate"] <= .1][
    ["st_share","sc_share", "ptg_share", "inc_5k_plus_share", "ed_prim_share","land_own_share", "crime_rate"]].dropna()
```

```python
def norm(col):
    return (col - col.mean())/col.std()

```

```python
X = sm.add_constant(analysis_df[["sc_share", "st_share", "ptg_share", "inc_5k_plus_share", "ed_prim_share","land_own_share"]].apply(norm, axis=0))
y = norm(analysis_df["crime_rate"])
```

```python

ols = sm.OLS(y, X)

results = ols.fit()

results.summary()
```

```python
sns.scatterplot(data=analysis_df, y="crime_rate", x="sc_share")
```
