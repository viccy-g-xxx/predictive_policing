```python
import pandas as pd
```

```python
shrid12_keys_path = '../data/shrug/shrug-shrid-keys-csv/shrid1_shrid2_key.csv'
shrid_dis_path = '../data/shrug/shrug-pc-keys-csv/shrid_pc11dist_key.csv'
districts = '../data/india-districts-census-2011.csv'
shrid_loc_names = '../data/shrug/shrug-shrid-keys-csv/shrid_loc_names.csv'

secc_rural_path = '../data/shrug/shrug-secc-mord-rural-csv/secc_rural_shrid.csv'
```

```python
shrid_names = pd.read_csv(shrid_loc_names)
shrid_names
```

```python
district_codes = pd.read_csv(districts)[["District code", "State name", "District name"]]
shrid12_keys = pd.read_csv(shrid12_keys_path)
shrid_district = pd.read_csv(shrid_dis_path)

shrid_keys = shrid12_keys.merge(shrid_district, how = 'inner', on = 'shrid2')
```

```python
shrid_df = shrid_keys.merge(shrid_names, on = 'shrid2')
```

```python
secc_rural = pd.read_csv(secc_rural_path)
```

```python
merged_df = shrid_df.merge(secc_rural, how = "inner", on = "shrid2")
```

```python
merged_df.groupby(["pc11_state_id", "pc11_district_id", "state_name", "district_name"])[
    [
        "st_share", 
        "sc_share",
        "ptg_share", 
        "tot_p", 
        "inc_5k_plus_share", 
        "ed_prim_share",
        "land_own_share",
        "disab_share",
        "scav_share",
        "house_type3"]].agg(
    {"st_share":"mean", 
     "sc_share":"mean", 
     "ptg_share":"mean", 
     "tot_p":'sum', 
     "inc_5k_plus_share":"mean",
     "disab_share":"mean",
     "scav_share":"mean",
     "house_type3":"mean",
    "ed_prim_share":"mean",
    "land_own_share":"mean"}).reset_index().to_csv("../data/district_case_comps.csv",index=False)
```
